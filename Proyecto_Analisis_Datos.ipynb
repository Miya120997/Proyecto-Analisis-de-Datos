# Introducción

El análisis de variables climáticas a lo largo del tiempo permite identificar patrones de cambio que pueden estar relacionados con fenómenos naturales como El Niño, La Niña o el cambio climático global [5]. En este contexto, el presente proyecto tiene como objetivo analizar datos climáticos históricos obtenidos del portal NASA POWER, comparando dos periodos específicos: 1997–1998 y 2017–2018, con el fin de identificar tendencias, correlaciones y posibles transformaciones en el comportamiento hidrológico.

El archivo de datos en formato Excel contiene registros diarios de variables hidrometeorológicas clave, entre las que se destacan:

-ETP (Evapotranspiración Potencial): cantidad máxima de agua que puede evaporarse bajo condiciones ideales de humedad del suelo, determinada principalmente por la energía solar disponible [1].

-ETR (Evapotranspiración Real): cantidad de agua que realmente se evapora y transpira desde el suelo y la vegetación, y que depende de la disponibilidad hídrica [3].

-LLU (Lluvia): precipitación diaria acumulada, expresada en milímetros [7].

-ESC (Escorrentía): agua superficial que fluye tras eventos de precipitación y que puede variar según la pendiente, vegetación y tipo de suelo [2].

PERC (Percolación): fracción del agua que se infiltra profundamente en el perfil del suelo, más allá de la zona radicular, y que contribuye a la recarga de acuíferos [4].

Estas variables están directamente relacionadas con el balance hídrico, un componente fundamental para la gestión sostenible del agua, la planificación agrícola y la evaluación de riesgos climáticos [8]. La comparación de los dos periodos busca identificar cambios o anomalías en la disponibilidad y dinámica del agua, especialmente frente a posibles impactos del cambio climático.

El análisis fue realizado utilizando herramientas de programación científica como Python y librerías especializadas como pandas, matplotlib y seaborn, que permiten estructurar, explorar y visualizar grandes volúmenes de datos de forma eficiente [6].

Este proyecto contribuye a la comprensión de las transformaciones climáticas en periodos separados por dos décadas, aportando una base analítica útil para futuras investigaciones, planificación ambiental o toma de decisiones en el campo de las geociencias.

# Objetivo

Realizar un análisis de un conjunto de datos de elección libre, preferible con algún enfoque en geociencias, utilizando las librerías de Python numpy, pandas, matplotlib, seaborn, entre otras, para identificar tendencias y observaciones importantes de los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuración de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)
```

```python
df1 = pd.read_csv("clean_1997_1998.csv", parse_dates=["Fecha"])
df2 = pd.read_csv("clean_2017_2018.csv", parse_dates=["Fecha"])

df1.head()
```

```python
print("Resumen estadístico de 1997-1998:")
display(df1.describe())

print("Resumen estadístico de 2017-2018:")
display(df2.describe())
```

## Distribución de variables climáticas por año

```python
for col in ['ETP', 'ETR', 'LLU', 'ESC', 'PERC']:
    if col in df1.columns and col in df2.columns:
        sns.kdeplot(df1[col], label=f"{col} 1997-98", fill=True)
        sns.kdeplot(df2[col], label=f"{col} 2017-18", fill=True)
        plt.title(f"Distribución de {col}")
        plt.legend()
        plt.show()
```

## Tendencias de variables climáticas a lo largo del tiempo

```python
for col in ['ETP', 'ETR', 'LLU']:
    if col in df1.columns:
        plt.plot(df1['Fecha'], df1[col], label='1997-98')
    if col in df2.columns:
        plt.plot(df2['Fecha'], df2[col], label='2017-18')
    plt.title(f"Evolución temporal de {col}")
    plt.legend()
    plt.xlabel("Fecha")
    plt.ylabel(col)
    plt.show()
```

```python
corr1 = df1.corr(numeric_only=True)
corr2 = df2.corr(numeric_only=True)

sns.heatmap(corr1, annot=True, cmap='coolwarm')
plt.title("Correlación 1997-1998")
plt.show()

sns.heatmap(corr2, annot=True, cmap='coolwarm')
plt.title("Correlación 2017-2018")
plt.show()
```

## Conclusiones

- Se observa una diferencia en las distribuciones de ETP y LLU entre los períodos 1997-1998 y 2017-2018.
- La variabilidad interanual puede estar relacionada con fenómenos como El Niño o el cambio climático.
- Las visualizaciones permiten detectar diferencias notables en los patrones de precipitación y escorrentía.
- El análisis de correlación muestra relaciones distintas entre las variables en ambos periodos.
